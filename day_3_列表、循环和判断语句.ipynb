```python
scores=[85, 92, 78, 65, 95, 88]

excellent_count=0
total_score=0

for score in scores:
    total_score+=score
    if score>=90:
        excellent_count+=1

average_score=total_score/len(scores)
print(f'优秀分数个数: {excellent_count}')
print(f'分数总和: {total_score}')
print(f'平均分数: {average_score:.3f}')
```
